###2.1 Temporal Features

```python
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time'])
```

```python
df_train['hour'] = df_train['trans_date_trans_time'].dt.hour
df_train['dayofweek'] = df_train['trans_date_trans_time'].dt.dayofweek
df_train['month'] = df_train['trans_date_trans_time'].dt.month
df_train['year'] = df_train['trans_date_trans_time'].dt.year
```

###2.2 **Data Preprocessing Choices| Deleting Unnecessary columnss**  
- Removed PII columns (`first`, `last`, `street`) for privacy compliance  
- Kept `cc_num` for behavioral pattern analysis  
- Used Haversine formula for distance calculation (accuracy ±1%)

```python
df_train = df_train.drop(columns = 'Unnamed: 0')
df_train = df_train.drop(columns = ['first','last','street','city','trans_num','unix_time'])
```

###2.3 **Geospatial Feature:** Distance from Home






Calculated distance between customer (lat, long) and merchant (merch_lat, merch_long) using the Haversine formula

```python
from math import radians, sin, cos, sqrt, atan2

# Calculate distance between cardholder and merchant using Haversine formula
def haversine_distance(row):
    lat1, lon1 = radians(row['lat']), radians(row['long'])
    lat2, lon2 = radians(row['merch_lat']), radians(row['merch_long'])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    r = 6371  # Earth's radius in km
    return r * c

df_train['distance'] = df_train.apply(haversine_distance, axis=1)
df_train.head()
```

###2.3 Behavioral Features

Aggregated customer-level metrics using cc_num:





Transaction count, average amount, fraud rate

```python
qism = df_train[['cc_num', 'amt', 'is_fraud']]
cust_agg = qism.groupby('cc_num').agg(cust_tranc_count=('amt','count'),
    cust_avg_amount=('amt','mean'),
    cust_fraud_rate=('is_fraud','mean')
).reset_index()
df_train
```

```python
df_train = df_train.merge(cust_agg, how='left', on='cc_num')
```

```python
df_train.head()
```

merchant-level summary of transaction counts, average amounts, and fraud rates

```python
mercant_qism = df_train[['merchant','amt','is_fraud']]
mercant_agg = mercant_qism.groupby('merchant').agg(mercant_tranc_count=('amt','count'),
    mercant_avg_amount=('amt','mean'),
    mercant_fraud_rate=('is_fraud','mean'))
mercant_agg
```

```python
df_train = df_train.merge(mercant_agg, how='left', on='merchant')
```

Derived customer age from dob and transaction date

```python
df_train['dob'] = pd.to_datetime(df_train['dob'])
df_train['age'] = (df_train['trans_date_trans_time'] - df_train['dob']).dt.days // 365
```

### 4.1 Key Findings

- **Categorical Features**:
  - Categories like `misc_net` and states like `NY` are prominent.

- **Time Patterns**:
  - Fraud peaks late at night (0-3 AM, 10-11 PM) and slightly on weekends.

- **Class Imbalance**:
  - Severe (~1% fraud), requiring techniques like oversampling or weighted models.


Recommendations
Preprocessing: Log-transform amt, encode category and state.
Monitoring: Increase scrutiny during high-risk hours (0-3 AM, 10-11 PM).
Next Steps: Prepare for modeling with imbalance correction (e.g., SMOTE) and evaluate using precision-recall metrics.

```python
from sklearn.preprocessing import StandardScaler

df_transformed = df_train.copy()

amt_cap = df_transformed['amt'].quantile(0.99)
df_transformed['amt'] = np.log1p(np.minimum(df_transformed['amt'], amt_cap))

#city_pop: log1p
df_transformed['city_pop'] = np.log1p(df_transformed['city_pop'])



# age: StandardScaler
scaler_age = StandardScaler()
df_transformed['age_scaled'] = scaler_age.fit_transform(df_transformed[['age']])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.histplot(df_train['amt'], bins=50, kde=True, ax=axes[0,0]).set_title('Original amt')
sns.histplot(df_transformed['amt'], bins=50, kde=True, ax=axes[0,1]).set_title('Transformed amt')

sns.histplot(df_train['city_pop'], bins=50, kde=True, ax=axes[1,0]).set_title('Original city_pop')
sns.histplot(df_transformed['city_pop'], bins=50, kde=True, ax=axes[1,1]).set_title('Transformed city_pop')

plt.tight_layout()
plt.show()
```
